# recoder les variables qualitatives

```python
#recoder

def  recod_qual(serie):
    return serie.astype('category').cat.codes
def recod_data_qual(dfs):
    for col in dfs.select_dtypes('object').columns:
        dfs[col] =  recod_qual(dfs[col])
    return dfs
    
df= recod_data_qual(df)
df
```

# normalisation des variables quantitatives

```python


def normlise_v_quant(df):
    #for i in df.select_dtypes('int64'):
    #  if 
    for i in df.columns:
        if(i in df.select_dtypes('object').columns)==False:
            df[i]=df[i]/df[i].max()
    return df
normlise_v_quant(df)
```
